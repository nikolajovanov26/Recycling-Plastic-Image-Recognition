```python
import os
from PIL import Image
```

```python
import glob
image_list = []
loc='test/Balans/'
ext='*.png'
loc1="".join((loc, ext))
for filename in glob.glob(loc1):
    base=os.path.basename(filename)
    im=Image.open(filename)
    str = os.path.splitext(base)[0]
    im.save(loc+str+".jpg", quality=20)
    image_list.append(im)
```
